```python
import os
directory_in_str = "./Palettes"
directory = os.fsencode(directory_in_str)
dirList = sorted(os.listdir(directory), key=len)
frames = []
for file in dirList:
    filename = os.fsdecode(file)
    if filename.endswith(".png"): 
        #print(filename)
        img = Image.open(directory_in_str + "/" + filename)
        frames.append(img)
    else:
        continue
```

```python
frames[0]
```

```python
# Save frames as an animated GIF
frames[0].save(
    'Gifs/animated.gif',
    save_all=True,
    append_images=frames[1:],
    duration=60,
    loop=0
)
```

```python

# Save the GIF with optimization
frames[0].save(
    'Gifs/optimized_animated.gif',
    save_all=True,
    append_images=frames[1:],
    duration=60,
    loop=0,
    optimize=True
)

```
