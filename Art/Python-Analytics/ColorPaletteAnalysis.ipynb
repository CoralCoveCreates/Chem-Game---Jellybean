```python
import numpy as np
import math
import PIL
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
import copy
```

```python
def countColors (img,colorCountsm):
    
    for x in range(img.width):
        for y in range(img.height):
            (r,g,b,a) = img.getpixel((x,y))
            colorCountsm[r,g,b] += 1 
            #if (r==131) and (g==128) and (b==137):
            #    print("I was here")
    return colorCountsm
```

```python
import os
directory_in_str = "../Color Palette Dataset"
directory = os.fsencode(directory_in_str)

colorCountsm = np.zeros((256,256,256))
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".png"): 
        #print(filename)
        img = Image.open(directory_in_str + "/" + filename)
        if img.mode == 'RGBA' or 'rgba':
            colorCountsm = countColors(img,colorCountsm)
        else:
            print(filename)
    else:
        continue
```

```python
def makeColorl (colorCountsm_, lowerbound_, upperbound_= colorCountsm.max()+1):
    commonColorsl = []
    for x in range(colorCountsm_.shape[0]):
        for y in range(colorCountsm_.shape[1]):
            for z in range(colorCountsm_.shape[2]):
                if colorCountsm_[x,y,z] > lowerbound_ and colorCountsm_[x,y,z] < upperbound_:
                    commonColorsl.append((x,y,z))

    print("Number of colors in common colors list", len(commonColorsl))

    refinedCommonColorsl = []
    for color in commonColorsl:
        if color[0] > 5 and color[1] > 5 and color[2]>5:
            refinedCommonColorsl.append(color)

    print("Number of colors in refined common colors list", len(refinedCommonColorsl))
    print("Roughly how many blocks wide and blocks tall does the palette need to be? ", np.round(math.sqrt(len(refinedCommonColorsl)),2))
    return refinedCommonColorsl


def makePaletteBitmapImagem (nBlocksWide: int, nBlocksHigh: int, colorl:list, blockSize=20):

    paletteBitmapImagem = np.zeros((nBlocksWide*blockSize,nBlocksHigh*blockSize,3))
    nBlocksInImage = nBlocksWide*nBlocksHigh
    nColors = len(colorl)

    if nBlocksInImage < nColors:
         raise ValueError("Too many colors and not enough blocks for them in the image ")
    else:
        i_=0
        for x in range(nBlocksWide):
            for y in range(nBlocksHigh):
                    if i_<len(colorl):
                        pbi_startx = x*blockSize
                        pbi_endx = (x*blockSize)+blockSize
                        pbi_starty = y*blockSize 
                        pbi_endy = (y*blockSize)+blockSize
                        paletteBitmapImagem[pbi_startx:pbi_endx, pbi_starty:pbi_endy, :] =  colorl[i_]
                    i_+=1

        paletteBitmapImagem = paletteBitmapImagem.astype(int)
        return paletteBitmapImagem
```

```python
colorl_gt5000 = makeColorl(colorCountsm_=colorCountsm, lowerbound_=5000)
```

```python
try:
    paletteBitmapImagem_0 = makePaletteBitmapImagem(nBlocksWide=10, nBlocksHigh=9, colorl=colorl_gt5000)
except ValueError as e:
    print(e)
plt.imshow(paletteBitmapImagem_0)
plt.show()
```
